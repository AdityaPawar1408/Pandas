# Arithmetic Operation

```python
var=pd.DataFrame({"A":[1,3,4,5],"B":[4,5,7,8]})
var["C"]=var["A"]+var["B"]
var["D"]=var["B"]-var["A"]
var["E"]=var["A"]*var["B"]
var["F"]=var["A"]/var["B"]
var
```

```python
import pandas as pd
var1=pd.DataFrame({"A":[10,30,40,50],"B":[11,20,33,44]})
var1["python"]=var1["A"]<=30
var1["python_1"]=var1["B"]>=22
var1["python_2"]=var1["B"]//11

var1
```
